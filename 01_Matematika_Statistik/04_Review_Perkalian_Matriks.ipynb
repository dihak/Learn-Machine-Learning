## Definisi Perkalian Matriks

Jika A adalah matriks berukuran m×n dan B adalah matriks berukuran n×p, maka perkalian A×B akan menghasilkan matriks C berukuran m×p. Syarat perkalian matriks adalah jumlah kolom pada matriks pertama harus sama dengan jumlah baris pada matriks kedua.

Elemen C pada baris i dan kolom j dihitung dengan:

$$c_{ij} = \sum_{k=1}^{n} a_{ik} \times b_{kj}$$

Di mana:
- $a_{ik}$ adalah elemen matriks A pada baris i, kolom k
- $b_{kj}$ adalah elemen matriks B pada baris k, kolom j

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import time

# Membuat matriks contoh
A = np.array([[1, 2, 3],
              [4, 5, 6]])  # Matriks 2×3

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])  # Matriks 3×2

print("Matriks A (2×3):")
print(A)
print("\nMatriks B (3×2):")
print(B)

# Perkalian matriks
C = A @ B  # atau np.matmul(A, B) atau np.dot(A, B)
print("\nHasil perkalian A × B (2×2):")
print(C)

# Verifikasi dimensi hasil perkalian
print(f"\nDimensi A: {A.shape}")
print(f"Dimensi B: {B.shape}")
print(f"Dimensi A×B: {C.shape}")
print(f"Dimensi hasil perkalian: {A.shape[0]}×{B.shape[1]}")
```

```python
# Perhitungan manual untuk elemen-elemen hasil
# c_00 = a_00*b_00 + a_01*b_10 + a_02*b_20 = 1*7 + 2*9 + 3*11 = 7 + 18 + 33 = 58
c_00 = A[0,0]*B[0,0] + A[0,1]*B[1,0] + A[0,2]*B[2,0]
print(f"C[0,0] = A[0,0]*B[0,0] + A[0,1]*B[1,0] + A[0,2]*B[2,0]")
print(f"       = {A[0,0]}*{B[0,0]} + {A[0,1]}*{B[1,0]} + {A[0,2]}*{B[2,0]}")
print(f"       = {A[0,0]*B[0,0]} + {A[0,1]*B[1,0]} + {A[0,2]*B[2,0]}")
print(f"       = {c_00}")
print(f"Verifikasi dengan hasil NumPy: C[0,0] = {C[0,0]}")

# c_01 = a_00*b_01 + a_01*b_11 + a_02*b_21 = 1*8 + 2*10 + 3*12 = 8 + 20 + 36 = 64
c_01 = A[0,0]*B[0,1] + A[0,1]*B[1,1] + A[0,2]*B[2,1]
print(f"\nC[0,1] = A[0,0]*B[0,1] + A[0,1]*B[1,1] + A[0,2]*B[2,1]")
print(f"       = {A[0,0]}*{B[0,1]} + {A[0,1]}*{B[1,1]} + {A[0,2]}*{B[2,1]}")
print(f"       = {A[0,0]*B[0,1]} + {A[0,1]*B[1,1]} + {A[0,2]*B[2,1]}")
print(f"       = {c_01}")
print(f"Verifikasi dengan hasil NumPy: C[0,1] = {C[0,1]}")

# c_10 = a_10*b_00 + a_11*b_10 + a_12*b_20 = 4*7 + 5*9 + 6*11 = 28 + 45 + 66 = 139
c_10 = A[1,0]*B[0,0] + A[1,1]*B[1,0] + A[1,2]*B[2,0]
print(f"\nC[1,0] = A[1,0]*B[0,0] + A[1,1]*B[1,0] + A[1,2]*B[2,0]")
print(f"       = {A[1,0]}*{B[0,0]} + {A[1,1]}*{B[1,0]} + {A[1,2]}*{B[2,0]}")
print(f"       = {A[1,0]*B[0,0]} + {A[1,1]*B[1,0]} + {A[1,2]*B[2,0]}")
print(f"       = {c_10}")
print(f"Verifikasi dengan hasil NumPy: C[1,0] = {C[1,0]}")

# c_11 = a_10*b_01 + a_11*b_11 + a_12*b_21 = 4*8 + 5*10 + 6*12 = 32 + 50 + 72 = 154
c_11 = A[1,0]*B[0,1] + A[1,1]*B[1,1] + A[1,2]*B[2,1]
print(f"\nC[1,1] = A[1,0]*B[0,1] + A[1,1]*B[1,1] + A[1,2]*B[2,1]")
print(f"       = {A[1,0]}*{B[0,1]} + {A[1,1]}*{B[1,1]} + {A[1,2]}*{B[2,1]}")
print(f"       = {A[1,0]*B[0,1]} + {A[1,1]*B[1,1]} + {A[1,2]*B[2,1]}")
print(f"       = {c_11}")
print(f"Verifikasi dengan hasil NumPy: C[1,1] = {C[1,1]}")

# Visualisasi perhitungan perkalian matriks
def visualize_matrix_multiplication(A, B):
    """Visualisasi perkalian matriks untuk matriks 2D"""
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    
    # Plot matriks A
    axes[0].matshow(A, cmap='Blues')
    axes[0].set_title('Matriks A')
    for i in range(A.shape[0]):
        for j in range(A.shape[1]):
            axes[0].text(j, i, f'{A[i, j]}', va='center', ha='center', fontsize=15)
    
    # Plot matriks B
    axes[1].matshow(B, cmap='Greens')
    axes[1].set_title('Matriks B')
    for i in range(B.shape[0]):
        for j in range(B.shape[1]):
            axes[1].text(j, i, f'{B[i, j]}', va='center', ha='center', fontsize=15)
    
    # Plot hasil A×B
    C = A @ B
    axes[2].matshow(C, cmap='Reds')
    axes[2].set_title('Matriks C = A×B')
    for i in range(C.shape[0]):
        for j in range(C.shape[1]):
            axes[2].text(j, i, f'{C[i, j]}', va='center', ha='center', fontsize=15)
    
    plt.tight_layout()
    plt.show()

# Visualisasi perkalian matriks
visualize_matrix_multiplication(A, B)
```

## Implementasi Perkalian Matriks dari Awal

Untuk memahami perkalian matriks lebih dalam, mari kita implementasikan operasi perkalian matriks dari awal (dari scratch) dan bandingkan dengan implementasi NumPy.

```python
def matrix_multiply(A, B):
    """
    Implementasi perkalian matriks dari awal (tanpa NumPy)
    
    Parameters:
    A (list of lists): Matriks pertama
    B (list of lists): Matriks kedua
    
    Returns:
    list of lists: Hasil perkalian A×B
    """
    # Verifikasi dimensi
    if len(A[0]) != len(B):
        raise ValueError("Dimensi matriks tidak sesuai untuk perkalian")
    
    # Inisialisasi matriks hasil
    m = len(A)        # Jumlah baris A
    n = len(A[0])     # Jumlah kolom A / Jumlah baris B
    p = len(B[0])     # Jumlah kolom B
    
    C = [[0 for _ in range(p)] for _ in range(m)]
    
    # Lakukan perkalian
    for i in range(m):
        for j in range(p):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    
    return C

# Uji implementasi perkalian matriks dari awal
A_list = [[1, 2, 3], [4, 5, 6]]
B_list = [[7, 8], [9, 10], [11, 12]]

result_manual = matrix_multiply(A_list, B_list)
print("Hasil perkalian matriks (implementasi manual):")
for row in result_manual:
    print(row)

print("\nHasil perkalian matriks (NumPy):")
print(A @ B)

# Verifikasi kesamaan hasil
result_np = (A @ B).tolist()
print("\nApakah hasil sama?", result_manual == result_np)

# Perbandingan performa
def compare_performance(size=500):
    """Membandingkan performa perkalian matriks manual vs NumPy"""
    # Buat matriks acak
    A_np = np.random.rand(size, size)
    B_np = np.random.rand(size, size)
    
    # Konversi ke list untuk implementasi manual
    A_list = A_np.tolist()
    B_list = B_np.tolist()
    
    # Ukur waktu untuk NumPy
    start_time = time.time()
    _ = A_np @ B_np
    np_time = time.time() - start_time
    print(f"Waktu NumPy: {np_time:.6f} detik")
    
    # Ukur waktu untuk implementasi manual (hanya untuk ukuran kecil)
    if size <= 100:
        start_time = time.time()
        _ = matrix_multiply(A_list, B_list)
        manual_time = time.time() - start_time
        print(f"Waktu manual: {manual_time:.6f} detik")
        print(f"NumPy lebih cepat {manual_time/np_time:.1f}x")
    else:
        print("Implementasi manual terlalu lambat untuk ukuran matriks yang besar")

# Bandingkan performa untuk matriks kecil
print("\nPerbandingan performa untuk matriks 50×50:")
compare_performance(50)

# Bandingkan performa untuk matriks besar
print("\nPerforma NumPy untuk matriks 500×500:")
compare_performance(500)
```

## Sifat-sifat Perkalian Matriks

Perkalian matriks memiliki beberapa sifat penting yang perlu dipahami:

1. **Tidak komutatif**: Umumnya, A×B ≠ B×A. Urutan perkalian penting.
   
2. **Asosiatif**: (A×B)×C = A×(B×C). Ini memungkinkan kita melakukan perkalian bertahap.
   
3. **Distributif terhadap penjumlahan**: A×(B+C) = A×B + A×C dan (A+B)×C = A×C + B×C.
   
4. **Perkalian dengan matriks identitas**: A×I = I×A = A, di mana I adalah matriks identitas.
   
5. **Perkalian dengan matriks transpose**: (A×B)^T = B^T × A^T. Perhatikan pembalikan urutan.

```python
# Uji sifat-sifat perkalian matriks

# Buat beberapa matriks contoh
A = np.array([[1, 2], 
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])
C = np.array([[9, 10],
              [11, 12]])

# 1. Sifat tidak komutatif: A×B ≠ B×A
AB = A @ B
BA = B @ A
print("1. Sifat tidak komutatif: A×B ≠ B×A")
print("A×B =")
print(AB)
print("\nB×A =")
print(BA)
print("\nApakah A×B = B×A?", np.array_equal(AB, BA))

# 2. Sifat asosiatif: (A×B)×C = A×(B×C)
AB_C = (A @ B) @ C
A_BC = A @ (B @ C)
print("\n\n2. Sifat asosiatif: (A×B)×C = A×(B×C)")
print("(A×B)×C =")
print(AB_C)
print("\nA×(B×C) =")
print(A_BC)
print("\nApakah (A×B)×C = A×(B×C)?", np.array_equal(AB_C, A_BC))

# 3. Sifat distributif: A×(B+C) = A×B + A×C
B_plus_C = B + C
A_BC_sum = A @ B_plus_C
AB_plus_AC = (A @ B) + (A @ C)
print("\n\n3. Sifat distributif: A×(B+C) = A×B + A×C")
print("A×(B+C) =")
print(A_BC_sum)
print("\nA×B + A×C =")
print(AB_plus_AC)
print("\nApakah A×(B+C) = A×B + A×C?", np.array_equal(A_BC_sum, AB_plus_AC))

# 4. Perkalian dengan matriks identitas
I = np.eye(2)  # Matriks identitas 2×2
AI = A @ I
IA = I @ A
print("\n\n4. Perkalian dengan matriks identitas: A×I = I×A = A")
print("A =")
print(A)
print("\nI =")
print(I)
print("\nA×I =")
print(AI)
print("\nI×A =")
print(IA)
print("\nApakah A×I = A?", np.array_equal(AI, A))
print("Apakah I×A = A?", np.array_equal(IA, A))

# 5. Perkalian dengan matriks transpose
AB_transpose = (A @ B).T
BT_AT = B.T @ A.T
print("\n\n5. Perkalian dengan matriks transpose: (A×B)^T = B^T × A^T")
print("(A×B)^T =")
print(AB_transpose)
print("\nB^T × A^T =")
print(BT_AT)
print("\nApakah (A×B)^T = B^T × A^T?", np.array_equal(AB_transpose, BT_AT))
```

## Jenis-jenis Perkalian Matriks

Selain perkalian matriks standar, ada beberapa jenis perkalian matriks lain yang sering digunakan dalam machine learning:

### 1. Perkalian Elemen per Elemen (Hadamard Product)

Perkalian Hadamard (elemen per elemen) antara dua matriks A dan B dengan dimensi yang sama, dilambangkan A⊙B, didefinisikan sebagai:

$$(A \odot B)_{ij} = A_{ij} \times B_{ij}$$

Dalam NumPy, ini dilakukan dengan operator `*`.

### 2. Perkalian Outer (Outer Product)

Perkalian outer antara vektor a dan b menghasilkan matriks C di mana:

$$C_{ij} = a_i \times b_j$$

### 3. Perkalian Kronecker

Perkalian Kronecker, dilambangkan A⊗B, adalah generalisasi dari perkalian outer. Jika A adalah matriks m×n dan B adalah matriks p×q, maka A⊗B adalah matriks mp×nq.

```python
# 1. Perkalian Hadamard (elemen per elemen)
A = np.array([[1, 2], 
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

hadamard = A * B  # Perkalian elemen per elemen
print("1. Perkalian Hadamard (elemen per elemen)")
print("A =")
print(A)
print("\nB =")
print(B)
print("\nA ⊙ B =")
print(hadamard)

# 2. Perkalian Outer
a = np.array([1, 2, 3])
b = np.array([4, 5])

outer_product = np.outer(a, b)
print("\n\n2. Perkalian Outer")
print("a =", a)
print("b =", b)
print("\na ⊗ b =")
print(outer_product)

# Manual calculation
print("\nPerhitungan manual:")
for i in range(len(a)):
    for j in range(len(b)):
        print(f"a[{i}] * b[{j}] = {a[i]} * {b[j]} = {a[i]*b[j]}")

# 3. Perkalian Kronecker
A = np.array([[1, 2], 
              [3, 4]])
B = np.array([[0, 5],
              [6, 7]])

kronecker = np.kron(A, B)
print("\n\n3. Perkalian Kronecker")
print("A =")
print(A)
print("\nB =")
print(B)
print("\nA ⊗ B =")
print(kronecker)

# Visualisasi
def visualize_hadamard_product(A, B):
    """Visualisasi perkalian Hadamard (elemen per elemen)"""
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    
    # Plot matriks A
    axes[0].matshow(A, cmap='Blues')
    axes[0].set_title('Matriks A')
    for i in range(A.shape[0]):
        for j in range(A.shape[1]):
            axes[0].text(j, i, f'{A[i, j]}', va='center', ha='center', fontsize=15)
    
    # Plot matriks B
    axes[1].matshow(B, cmap='Greens')
    axes[1].set_title('Matriks B')
    for i in range(B.shape[0]):
        for j in range(B.shape[1]):
            axes[1].text(j, i, f'{B[i, j]}', va='center', ha='center', fontsize=15)
    
    # Plot hasil A*B (hadamard)
    C = A * B
    axes[2].matshow(C, cmap='Reds')
    axes[2].set_title('Matriks C = A ⊙ B')
    for i in range(C.shape[0]):
        for j in range(C.shape[1]):
            axes[2].text(j, i, f'{C[i, j]}', va='center', ha='center', fontsize=15)
    
    plt.tight_layout()
    plt.show()

# Visualisasi perkalian Hadamard
A = np.array([[1, 2], 
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])
visualize_hadamard_product(A, B)

# Visualisasi perkalian outer
def visualize_outer_product(a, b):
    """Visualisasi perkalian outer antara dua vektor"""
    outer = np.outer(a, b)
    
    fig, ax = plt.subplots(figsize=(10, 8))
    im = ax.imshow(outer, cmap='viridis')
    
    # Tambahkan nilai ke setiap sel
    for i in range(len(a)):
        for j in range(len(b)):
            text = ax.text(j, i, f'{a[i]}×{b[j]}={outer[i, j]}', 
                          ha="center", va="center", color="w", fontsize=12)
    
    ax.set_xticks(range(len(b)))
    ax.set_xticklabels([f'b[{i}]={val}' for i, val in enumerate(b)])
    ax.set_yticks(range(len(a)))
    ax.set_yticklabels([f'a[{i}]={val}' for i, val in enumerate(a)])
    
    plt.colorbar(im)
    plt.title('Visualisasi Perkalian Outer (a ⊗ b)')
    plt.tight_layout()
    plt.show()

# Visualisasi perkalian outer
a = np.array([1, 2, 3])
b = np.array([4, 5])
visualize_outer_product(a, b)
```

## Perkalian Matriks di Machine Learning

Perkalian matriks memiliki banyak aplikasi dalam machine learning. Mari kita lihat beberapa contoh:

### 1. Transformasi Linear

Dalam transformasi linear, kita mengalikan matriks transformasi dengan vektor untuk mendapatkan vektor yang ditransformasi. Secara umum, jika A adalah matriks transformasi dan x adalah vektor, maka A×x adalah hasil transformasi.

### 2. Regresi Linear

Dalam regresi linear, kita mencari koefisien β yang meminimalkan error:

$$\hat{\beta} = (X^T X)^{-1} X^T y$$

Di mana X adalah matriks data, X^T adalah transpose X, dan y adalah vektor target.

### 3. Neural Networks

Dalam jaringan saraf, perkalian matriks digunakan untuk menghitung hasil dot product antara bobot dan input:

$$z = W \times x + b$$

Di mana W adalah matriks bobot, x adalah vektor input, dan b adalah vektor bias.

```python
# 1. Transformasi Linear

# Buat data 2D acak
np.random.seed(42)
points = np.random.randn(30, 2)

# Matriks transformasi untuk rotasi 45 derajat
theta = np.pi/4  # 45 derajat dalam radian
rotation_matrix = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta), np.cos(theta)]
])

# Aplikasikan transformasi
transformed_points = points @ rotation_matrix.T  # Transpose karena kita ingin mengalikan dari kanan

# Visualisasi
plt.figure(figsize=(10, 10))
plt.scatter(points[:, 0], points[:, 1], label='Original', alpha=0.7)
plt.scatter(transformed_points[:, 0], transformed_points[:, 1], label='Rotated', alpha=0.7)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.grid(True)
plt.axis('equal')
plt.title('Transformasi Linear: Rotasi 45 Derajat')
plt.legend()
plt.show()

# 2. Regresi Linear

# Buat data sintetis untuk regresi
np.random.seed(42)
X = np.random.rand(100, 2)  # 100 sampel, 2 fitur
true_beta = np.array([3, 1.5])  # Koefisien sebenarnya
y = X @ true_beta + np.random.randn(100) * 0.5  # Target dengan noise

# Tambahkan kolom untuk intercept
X_design = np.column_stack((np.ones(X.shape[0]), X))

# Hitung koefisien regresi menggunakan rumus closed-form
# β = (X^T X)^(-1) X^T y
X_transpose_X = X_design.T @ X_design
X_transpose_y = X_design.T @ y
beta = np.linalg.inv(X_transpose_X) @ X_transpose_y

print("Regresi Linear:")
print(f"Koefisien sebenarnya: Intercept=0, Beta1={true_beta[0]}, Beta2={true_beta[1]}")
print(f"Koefisien hasil regresi: Intercept={beta[0]:.4f}, Beta1={beta[1]:.4f}, Beta2={beta[2]:.4f}")

# 3. Neural Network (contoh sederhana)

def sigmoid(x):
    """Fungsi aktivasi sigmoid"""
    return 1 / (1 + np.exp(-x))

# Buat data untuk binary classification
np.random.seed(42)
X_nn = np.random.randn(100, 2)  # 100 sampel, 2 fitur
y_nn = (X_nn[:, 0] + X_nn[:, 1] > 0).astype(int)  # Label positif jika x1 + x2 > 0

# Model neural network sederhana (perceptron)
input_size = 2
hidden_size = 3
output_size = 1

# Inisialisasi bobot acak
W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros(hidden_size)
W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros(output_size)

# Forward pass
# Hidden layer: z1 = X × W1 + b1
z1 = X_nn @ W1 + b1
a1 = sigmoid(z1)  # Aktivasi

# Output layer: z2 = a1 × W2 + b2
z2 = a1 @ W2 + b2
a2 = sigmoid(z2)  # Aktivasi final

# Prediksi (0 atau 1)
predictions = (a2 > 0.5).astype(int).squeeze()

# Akurasi
accuracy = np.mean(predictions == y_nn)
print(f"\nNeural Network (Perceptron) Accuracy: {accuracy:.2f}")

# Visualisasi decision boundary
def plot_decision_boundary(X, y, W1, b1, W2, b2):
    """Plot decision boundary dari neural network sederhana"""
    # Buat grid
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    
    # Evaluasi model pada grid
    grid = np.c_[xx.ravel(), yy.ravel()]
    z1_grid = grid @ W1 + b1
    a1_grid = sigmoid(z1_grid)
    z2_grid = a1_grid @ W2 + b2
    a2_grid = sigmoid(z2_grid)
    grid_predictions = (a2_grid > 0.5).astype(int).reshape(xx.shape)
    
    # Plot
    plt.figure(figsize=(10, 8))
    plt.contourf(xx, yy, grid_predictions, alpha=0.3, cmap='viridis')
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', 
                edgecolor='k', s=100, alpha=0.7)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('Neural Network Decision Boundary')
    plt.colorbar()
    plt.tight_layout()
    plt.show()

# Plot decision boundary
plot_decision_boundary(X_nn, y_nn, W1, b1, W2, b2)
```

## Optimasi Perkalian Matriks

Perkalian matriks dapat menjadi operasi yang mahal secara komputasi, terutama untuk matriks besar. Ada beberapa teknik optimasi:

### 1. Algoritma Optimasi

- **Algoritma Strassen**: Mengurangi kompleksitas dari O(n³) menjadi O(n^2.807) dengan membagi matriks menjadi blok.
- **Algoritma Coppersmith-Winograd**: Lebih lanjut mengurangi kompleksitas tetapi jarang digunakan dalam praktik.

### 2. Optimasi Hardware

- **Paralelisasi**: Menggunakan multiple threads atau cores.
- **Komputasi GPU**: Menggunakan GPU untuk perkalian matriks paralel (misalnya melalui CUDA).
- **Instruksi SIMD**: Menggunakan instruksi Single Instruction Multiple Data.

### 3. Optimasi Implementasi

- **Block matrix multiplication**: Meningkatkan locality cache.
- **Preprocessing**: Mengubah format matriks untuk perkalian yang lebih efisien.
- **Menghindari perkalian saat memungkinkan**: Misalnya dengan matriks sparse.

```python
# Implementasi optimasi perkalian matriks

# Implementasi algoritma Strassen (versi sederhana)
def strassen_multiply(A, B):
    """
    Implementasi sederhana dari algoritma perkalian matriks Strassen
    Bekerja untuk matriks persegi dengan ukuran pangkat 2 (2, 4, 8, ...)
    """
    n = A.shape[0]
    
    # Kasus basis: matriks 1x1
    if n == 1:
        return A * B
    
    # Bagi matriks menjadi empat submatriks
    mid = n // 2
    
    A11 = A[:mid, :mid]
    A12 = A[:mid, mid:]
    A21 = A[mid:, :mid]
    A22 = A[mid:, mid:]
    
    B11 = B[:mid, :mid]
    B12 = B[:mid, mid:]
    B21 = B[mid:, :mid]
    B22 = B[mid:, mid:]
    
    # Hitung tujuh produk rekursif (formula Strassen)
    P1 = strassen_multiply(A11 + A22, B11 + B22)
    P2 = strassen_multiply(A21 + A22, B11)
    P3 = strassen_multiply(A11, B12 - B22)
    P4 = strassen_multiply(A22, B21 - B11)
    P5 = strassen_multiply(A11 + A12, B22)
    P6 = strassen_multiply(A21 - A11, B11 + B12)
    P7 = strassen_multiply(A12 - A22, B21 + B22)
    
    # Hitung submatriks hasil
    C11 = P1 + P4 - P5 + P7
    C12 = P3 + P5
    C21 = P2 + P4
    C22 = P1 + P3 - P2 + P6
    
    # Gabungkan submatriks
    C = np.vstack([np.hstack([C11, C12]), np.hstack([C21, C22])])
    
    return C

# Tes performa untuk ukuran yang lebih kecil
def test_strassen_performance():
    # Ukuran matriks (pangkat 2)
    n = 32  # 32×32
    
    # Buat matriks persegi
    A = np.random.rand(n, n)
    B = np.random.rand(n, n)
    
    # Perkalian standar dengan NumPy
    start_time = time.time()
    C_numpy = A @ B
    numpy_time = time.time() - start_time
    print(f"Waktu perkalian NumPy: {numpy_time:.6f} detik")
    
    # Perkalian Strassen
    start_time = time.time()
    C_strassen = strassen_multiply(A, B)
    strassen_time = time.time() - start_time
    print(f"Waktu perkalian Strassen: {strassen_time:.6f} detik")
    
    # Verifikasi hasil
    print(f"Hasil sama? {np.allclose(C_numpy, C_strassen)}")
    
    # Perbandingan
    ratio = strassen_time / numpy_time
    print(f"Strassen/NumPy ratio: {ratio:.2f}x")
    
    return numpy_time, strassen_time

# Catatan: Untuk implementasi praktis, NumPy lebih cepat daripada
# implementasi Python murni, bahkan untuk algoritma yang lebih baik
# seperti Strassen, karena NumPy menggunakan implementasi C yang
# dioptimalkan dan instruksi SIMD.
print("Perbandingan performa Strassen vs NumPy:")
numpy_time, strassen_time = test_strassen_performance()

# Demonstrasi penggunaan berbagai backend untuk perkalian matriks
def benchmark_matrix_multiplication(sizes=[100, 500, 1000]):
    """Benchmark untuk perkalian matriks dengan berbagai ukuran"""
    results = []
    
    for size in sizes:
        A = np.random.rand(size, size)
        B = np.random.rand(size, size)
        
        # NumPy standard
        start_time = time.time()
        _ = A @ B
        numpy_time = time.time() - start_time
        
        results.append({
            'size': size,
            'numpy_time': numpy_time
        })
        
        print(f"Ukuran {size}×{size}: {numpy_time:.6f} detik")
    
    return results

print("\nBenchmark perkalian matriks dengan NumPy:")
results = benchmark_matrix_multiplication()

# Plot hasil benchmark
plt.figure(figsize=(10, 6))
sizes = [r['size'] for r in results]
times = [r['numpy_time'] for r in results]

plt.plot(sizes, times, 'o-', linewidth=2, markersize=10)
plt.xlabel('Ukuran Matriks (n×n)')
plt.ylabel('Waktu (detik)')
plt.title('Performa Perkalian Matriks')
plt.grid(True)
plt.show()

# Catatan tentang optimasi NumPy:
print("\nCatatan tentang optimasi perkalian matriks:")
print("1. NumPy menggunakan BLAS (Basic Linear Algebra Subprograms) untuk operasi matriks")
print("2. Pada banyak sistem, NumPy terhubung ke implementasi BLAS yang dioptimalkan seperti:")
print("   - OpenBLAS")
print("   - Intel MKL (Math Kernel Library)")
print("   - ATLAS (Automatically Tuned Linear Algebra Software)")
print("3. Optimasi lanjutan meliputi:")
print("   - Paralelisasi multi-thread")
print("   - Penggunaan AVX/SSE (instruksi SIMD)")
print("   - Optimasi cache locality")
print("   - Algoritma Strassen dan variannya untuk matriks besar")
```

```python
# Latihan: Implementasi transformasi linear 3D (rotasi) menggunakan perkalian matriks

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Buat data 3D acak
np.random.seed(42)
points_3d = np.random.randn(50, 3) * 2

# Matriks rotasi 3D
def rotation_matrix_3d(angle_x, angle_y, angle_z):
    """
    Buat matriks rotasi 3D dari sudut Euler (dalam radian)
    
    Parameters:
    angle_x, angle_y, angle_z: Sudut rotasi pada sumbu x, y, dan z
    
    Returns:
    numpy.ndarray: Matriks rotasi 3×3
    """
    # Rotasi pada sumbu X
    R_x = np.array([
        [1, 0, 0],
        [0, np.cos(angle_x), -np.sin(angle_x)],
        [0, np.sin(angle_x), np.cos(angle_x)]
    ])
    
    # Rotasi pada sumbu Y
    R_y = np.array([
        [np.cos(angle_y), 0, np.sin(angle_y)],
        [0, 1, 0],
        [-np.sin(angle_y), 0, np.cos(angle_y)]
    ])
    
    # Rotasi pada sumbu Z
    R_z = np.array([
        [np.cos(angle_z), -np.sin(angle_z), 0],
        [np.sin(angle_z), np.cos(angle_z), 0],
        [0, 0, 1]
    ])
    
    # Gabungkan rotasi (urutan: z -> y -> x)
    R = R_x @ R_y @ R_z
    
    return R

# Buat matriks rotasi dengan sudut berbeda
rotation = rotation_matrix_3d(np.pi/6, np.pi/4, np.pi/3)  # 30°, 45°, 60°

# Aplikasikan transformasi
rotated_points = points_3d @ rotation

# Visualisasi
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Plot data asli
ax.scatter(points_3d[:, 0], points_3d[:, 1], points_3d[:, 2], 
           c='blue', marker='o', s=50, label='Original')

# Plot data yang telah dirotasi
ax.scatter(rotated_points[:, 0], rotated_points[:, 1], rotated_points[:, 2], 
           c='red', marker='^', s=50, label='Rotated')

# Tambahkan sumbu koordinat
ax.quiver(0, 0, 0, 2, 0, 0, color='r', arrow_length_ratio=0.1, label='X')
ax.quiver(0, 0, 0, 0, 2, 0, color='g', arrow_length_ratio=0.1, label='Y')
ax.quiver(0, 0, 0, 0, 0, 2, color='b', arrow_length_ratio=0.1, label='Z')

# Rotasi sumbu koordinat
rotated_x = rotation @ np.array([2, 0, 0])
rotated_y = rotation @ np.array([0, 2, 0])
rotated_z = rotation @ np.array([0, 0, 2])

ax.quiver(0, 0, 0, rotated_x[0], rotated_x[1], rotated_x[2], 
         color='darkred', arrow_length_ratio=0.1, label='X rotated')
ax.quiver(0, 0, 0, rotated_y[0], rotated_y[1], rotated_y[2], 
         color='darkgreen', arrow_length_ratio=0.1, label='Y rotated')
ax.quiver(0, 0, 0, rotated_z[0], rotated_z[1], rotated_z[2], 
         color='darkblue', arrow_length_ratio=0.1, label='Z rotated')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([-3, 3])
ax.legend()
ax.set_title('Rotasi 3D dengan Perkalian Matriks')

plt.tight_layout()
plt.show()

# Tampilkan matriks rotasi
print("Matriks rotasi 3D:")
print(rotation)

# Verifikasi bahwa matriks rotasi adalah matriks ortogonal (R^T @ R = I)
RT_R = rotation.T @ rotation
print("\nVerifikasi ortogonalitas (R^T @ R):")
print(RT_R)
print("Apakah matriks rotasi ortogonal?", np.allclose(RT_R, np.eye(3), atol=1e-10))
```
