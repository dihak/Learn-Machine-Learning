# Review Konsep Vektor

Vektor adalah kuantitas matematis yang memiliki nilai (magnitude) dan arah. Dalam machine learning, vektor digunakan untuk merepresentasikan data, parameter, dan hasil komputasi.

## Definisi Vektor

Vektor dapat direpresentasikan sebagai array satu dimensi dengan n elemen. Vektor dengan n elemen disebut vektor n-dimensi, ditulis sebagai:

$$\vec{v} = [v_1, v_2, \ldots, v_n]$$

atau dalam notasi kolom:

$$\vec{v} = \begin{bmatrix} v_1 \\ v_2 \\ \vdots \\ v_n \end{bmatrix}$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat vektor menggunakan NumPy
v = np.array([1, 2, 3])
w = np.array([4, 5, 6])

print("Vektor v:", v)
print("Vektor w:", w)
print("Dimensi vektor v:", v.shape)

# Vektor baris vs vektor kolom dalam NumPy
v_row = np.array([[1, 2, 3]])  # vektor baris: bentuk (1, 3)
v_col = np.array([[1], [2], [3]])  # vektor kolom: bentuk (3, 1)

print("\nVektor baris v_row:")
print(v_row)
print("Dimensi v_row:", v_row.shape)

print("\nVektor kolom v_col:")
print(v_col)
print("Dimensi v_col:", v_col.shape)
```

## Operasi Dasar Vektor

### 1. Penjumlahan dan Pengurangan Vektor

Penjumlahan dan pengurangan vektor dilakukan per elemen. Operasi ini hanya dapat dilakukan jika kedua vektor memiliki dimensi yang sama.

Jika $\vec{v} = [v_1, v_2, \ldots, v_n]$ dan $\vec{w} = [w_1, w_2, \ldots, w_n]$, maka:

$$\vec{v} + \vec{w} = [v_1 + w_1, v_2 + w_2, \ldots, v_n + w_n]$$
$$\vec{v} - \vec{w} = [v_1 - w_1, v_2 - w_2, \ldots, v_n - w_n]$$

```python
# Penjumlahan vektor
v = np.array([1, 2, 3])
w = np.array([4, 5, 6])

v_plus_w = v + w
print("v + w =", v_plus_w)

# Pengurangan vektor
v_minus_w = v - w
print("v - w =", v_minus_w)

# Visualisasi penjumlahan vektor 2D
plt.figure(figsize=(8, 6))

# Vektor 2D untuk visualisasi
v_2d = np.array([2, 3])
w_2d = np.array([1, 2])
sum_2d = v_2d + w_2d

# Plot vektor v
plt.quiver(0, 0, v_2d[0], v_2d[1], angles='xy', scale_units='xy', scale=1, color='r', label='v = [2, 3]')

# Plot vektor w
plt.quiver(0, 0, w_2d[0], w_2d[1], angles='xy', scale_units='xy', scale=1, color='g', label='w = [1, 2]')

# Plot vektor v+w
plt.quiver(0, 0, sum_2d[0], sum_2d[1], angles='xy', scale_units='xy', scale=1, color='b', label='v + w = [3, 5]')

# Plot vektor w mulai dari titik akhir v (untuk menunjukkan sifat penjumlahan vektor)
plt.quiver(v_2d[0], v_2d[1], w_2d[0], w_2d[1], angles='xy', scale_units='xy', scale=1, color='g', linestyle='dotted')

plt.xlim(-1, 5)
plt.ylim(-1, 6)
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.title('Penjumlahan Vektor: v + w')
plt.legend()
plt.axis('equal')
plt.show()
```

### 2. Perkalian Skalar

Perkalian vektor dengan skalar mengalikan setiap elemen vektor dengan nilai skalar tersebut.

Jika $\vec{v} = [v_1, v_2, \ldots, v_n]$ dan $c$ adalah skalar, maka:

$$c\vec{v} = [cv_1, cv_2, \ldots, cv_n]$$

### 3. Dot Product (Perkalian Titik)

Dot product antara dua vektor dengan dimensi yang sama menghasilkan nilai skalar. Dot product antara $\vec{v}$ dan $\vec{w}$ dihitung dengan:

$$\vec{v} \cdot \vec{w} = \sum_{i=1}^{n} v_i w_i = v_1w_1 + v_2w_2 + \ldots + v_nw_n$$

Secara geometris, dot product $\vec{v} \cdot \vec{w} = ||\vec{v}|| \cdot ||\vec{w}|| \cdot \cos(\theta)$, di mana $\theta$ adalah sudut antara dua vektor.

```python
# Perkalian skalar
v = np.array([1, 2, 3])
c = 3

cv = c * v
print(f"{c} * v =", cv)

# Visualisasi perkalian skalar
plt.figure(figsize=(8, 6))

# Vektor 2D untuk visualisasi
v_2d = np.array([2, 1])
cv_2d = 2 * v_2d

# Plot vektor v
plt.quiver(0, 0, v_2d[0], v_2d[1], angles='xy', scale_units='xy', scale=1, color='r', label='v = [2, 1]')

# Plot vektor 2v
plt.quiver(0, 0, cv_2d[0], cv_2d[1], angles='xy', scale_units='xy', scale=1, color='b', label='2v = [4, 2]')

plt.xlim(-1, 5)
plt.ylim(-1, 3)
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.title('Perkalian Skalar: 2 * v')
plt.legend()
plt.axis('equal')
plt.show()

# Dot product
v = np.array([1, 2, 3])
w = np.array([4, 5, 6])

# Cara 1: Menggunakan np.dot
dot_product = np.dot(v, w)
print("\nDot product v · w =", dot_product)

# Cara 2: Menggunakan operator @
dot_product2 = v @ w
print("Verifikasi dengan operator '@':", dot_product2)

# Cara manual
manual_dot = sum(v[i] * w[i] for i in range(len(v)))
print("Verifikasi dengan perhitungan manual:", manual_dot)

# Visualisasi dot product dengan sudut
plt.figure(figsize=(8, 6))

# Vektor 2D untuk visualisasi
v_2d = np.array([3, 1])
w_2d = np.array([2, 2])

# Panjang (magnitude) dari vektor
v_magnitude = np.linalg.norm(v_2d)
w_magnitude = np.linalg.norm(w_2d)

# Dot product
dot_2d = np.dot(v_2d, w_2d)

# Cosinus sudut antara vektor
cos_theta = dot_2d / (v_magnitude * w_magnitude)
theta_rad = np.arccos(cos_theta)
theta_deg = np.degrees(theta_rad)

# Plot vektor
plt.quiver(0, 0, v_2d[0], v_2d[1], angles='xy', scale_units='xy', scale=1, color='r', label=f'v = {v_2d}')
plt.quiver(0, 0, w_2d[0], w_2d[1], angles='xy', scale_units='xy', scale=1, color='b', label=f'w = {w_2d}')

# Teks info
plt.text(1.5, 0.5, f'|v| = {v_magnitude:.2f}', fontsize=10)
plt.text(1.5, 0.2, f'|w| = {w_magnitude:.2f}', fontsize=10)
plt.text(1.5, -0.1, f'v·w = {dot_2d}', fontsize=10)
plt.text(1.5, -0.4, f'cos(θ) = {cos_theta:.2f}', fontsize=10)
plt.text(1.5, -0.7, f'θ = {theta_deg:.2f}°', fontsize=10)

plt.xlim(-1, 4)
plt.ylim(-1, 3)
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.title('Dot Product: v · w = |v|·|w|·cos(θ)')
plt.legend()
plt.axis('equal')
plt.show()
```

### 4. Cross Product (Perkalian Silang)

Cross product adalah operasi khusus untuk vektor 3 dimensi. Hasil dari cross product $\vec{v} \times \vec{w}$ adalah vektor baru yang tegak lurus terhadap kedua vektor $\vec{v}$ dan $\vec{w}$.

Jika $\vec{v} = [v_1, v_2, v_3]$ dan $\vec{w} = [w_1, w_2, w_3]$, maka:

$$\vec{v} \times \vec{w} = [v_2w_3 - v_3w_2, v_3w_1 - v_1w_3, v_1w_2 - v_2w_1]$$

### 5. Norma Vektor (Panjang atau Magnitude)

Norma atau panjang vektor $\vec{v}$ dihitung dengan:

$$||\vec{v}|| = \sqrt{\sum_{i=1}^{n} v_i^2} = \sqrt{v_1^2 + v_2^2 + \ldots + v_n^2}$$

Norma ini juga disebut sebagai norma Euclidean atau norma $L_2$.

```python
# Cross product
v = np.array([1, 2, 3])
w = np.array([4, 5, 6])

cross_product = np.cross(v, w)
print("Cross product v × w =", cross_product)

# Verifikasi bahwa hasil cross product tegak lurus dengan v dan w
dot_v_cross = np.dot(v, cross_product)
dot_w_cross = np.dot(w, cross_product)

print("v · (v × w) =", dot_v_cross)  # Seharusnya 0
print("w · (v × w) =", dot_w_cross)  # Seharusnya 0

# Visualisasi cross product
# Kita akan menggunakan vektor 3D yang lebih sederhana
from mpl_toolkits.mplot3d import Axes3D

v_3d = np.array([1, 0, 0])  # vektor i
w_3d = np.array([0, 1, 0])  # vektor j
cross_3d = np.cross(v_3d, w_3d)  # seharusnya vektor k = [0, 0, 1]

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot vektor-vektor
ax.quiver(0, 0, 0, v_3d[0], v_3d[1], v_3d[2], color='r', label='v = [1, 0, 0]')
ax.quiver(0, 0, 0, w_3d[0], w_3d[1], w_3d[2], color='g', label='w = [0, 1, 0]')
ax.quiver(0, 0, 0, cross_3d[0], cross_3d[1], cross_3d[2], color='b', label='v × w = [0, 0, 1]')

ax.set_xlim([-1, 1.5])
ax.set_ylim([-1, 1.5])
ax.set_zlim([-1, 1.5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
ax.set_title('Cross Product: v × w')
plt.show()

# Norma vektor (panjang atau magnitude)
v = np.array([3, 4])
w = np.array([1, 2, 3])

# Norma Euclidean (L2)
norm_v = np.linalg.norm(v)
norm_w = np.linalg.norm(w)

print("\nNorma v =", norm_v)  # Seharusnya 5 (teorema Pythagoras: 3²+4²=25, sqrt(25)=5)
print("Norma w =", norm_w)    # sqrt(1²+2²+3²) = sqrt(14) ≈ 3.74

# Verifikasi dengan perhitungan manual
norm_v_manual = np.sqrt(sum(v_i**2 for v_i in v))
print("Norma v (manual) =", norm_v_manual)

# Visualisasi norma vektor 2D
plt.figure(figsize=(8, 6))

# Plot vektor
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label=f'v = {v}')

# Plot garis bantu untuk menunjukkan komponen
plt.plot([0, v[0]], [0, 0], 'k--', alpha=0.3)
plt.plot([v[0], v[0]], [0, v[1]], 'k--', alpha=0.3)

plt.text(v[0]/2, -0.3, str(v[0]), fontsize=10)
plt.text(v[0]+0.2, v[1]/2, str(v[1]), fontsize=10)
plt.text(v[0]/2, v[1]/2, f"|v| = {norm_v}", fontsize=12)

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.title('Norma Vektor: |v| = √(v₁² + v₂²)')
plt.legend()
plt.axis('equal')
plt.show()
```

### 6. Unit Vector (Vektor Satuan)

Unit vector adalah vektor yang memiliki panjang (norma) 1. Vektor satuan dari vektor $\vec{v}$ dihitung dengan:

$$\hat{v} = \frac{\vec{v}}{||\vec{v}||}$$

Unit vector mempertahankan arah vektor asli tetapi mengubah panjangnya menjadi 1.

## Aplikasi Vektor dalam Machine Learning

Vektor memiliki banyak aplikasi dalam machine learning:

1. **Representasi Data**: Setiap sampel data sering direpresentasikan sebagai vektor fitur.

2. **Perhitungan Jarak**: Jarak antara dua sampel data dihitung menggunakan norma dari selisih vektor mereka.

3. **Proyeksi dan Dekomposisi**: Vektor digunakan dalam teknik seperti PCA untuk mereduksi dimensi data.

4. **Gradien**: Turunan parsial dalam optimasi fungsi multivariabel disusun dalam bentuk vektor gradien.

5. **Embedding**: Word embedding, image embedding, dan teknik representasi lainnya menghasilkan vektor.

Mari kita lihat beberapa contoh aplikasi vektor dalam konteks machine learning:

```python
# Unit Vector
v = np.array([3, 4])
norm_v = np.linalg.norm(v)

# Menghitung unit vector
unit_v = v / norm_v
print("v =", v)
print("Unit vector v̂ =", unit_v)
print("Norma dari unit vector |v̂| =", np.linalg.norm(unit_v))  # Seharusnya 1

# Visualisasi unit vector
plt.figure(figsize=(8, 6))

# Plot vektor asli
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label=f'v = {v}')

# Plot unit vector (dengan skala yang sama)
plt.quiver(0, 0, unit_v[0], unit_v[1], angles='xy', scale_units='xy', scale=1, color='b', label=f'v̂ = {unit_v.round(2)}')

plt.xlim(-1, 4)
plt.ylim(-1, 5)
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.title('Unit Vector: v̂ = v/|v|')
plt.legend()
plt.axis('equal')
plt.show()

# Aplikasi Vektor dalam Machine Learning

# 1. Representasi data sebagai vektor
# Contoh: Dua sampel dengan 4 fitur
sample1 = np.array([5.1, 3.5, 1.4, 0.2])  # Mungkin Iris setosa
sample2 = np.array([7.0, 3.2, 4.7, 1.4])  # Mungkin Iris versicolor

print("\nSampel 1:", sample1)
print("Sampel 2:", sample2)

# 2. Menghitung jarak antara dua sampel
# Jarak Euclidean
distance = np.linalg.norm(sample1 - sample2)
print("\nJarak Euclidean antara sampel 1 dan sampel 2:", distance)

# 3. Similarity (cosine similarity)
cos_sim = np.dot(sample1, sample2) / (np.linalg.norm(sample1) * np.linalg.norm(sample2))
print("Kesamaan cosinus antara sampel 1 dan sampel 2:", cos_sim)

# 4. Contoh proyeksi vektor
# Misalkan kita ingin memproyeksikan sample1 ke arah sample2
# Proyeksi dari v ke w adalah: proj_w(v) = (v·w / |w|²) * w
projection_magnitude = np.dot(sample1, sample2) / np.dot(sample2, sample2)
projection = projection_magnitude * sample2

print("\nProyeksi sampel 1 ke arah sampel 2:")
print(projection)
print("Panjang proyeksi:", np.linalg.norm(projection))

# 5. Visualisasi k-nearest neighbors sederhana dengan vektor 2D
# Buat beberapa contoh data 2D untuk kelas berbeda
np.random.seed(42)
class_a = np.random.randn(6, 2) + np.array([-2, -2])  # Cluster untuk kelas A
class_b = np.random.randn(6, 2) + np.array([2, 2])    # Cluster untuk kelas B
new_point = np.array([0, 0])  # Titik baru yang akan diklasifikasikan

plt.figure(figsize=(8, 6))
plt.scatter(class_a[:, 0], class_a[:, 1], color='red', label='Kelas A')
plt.scatter(class_b[:, 0], class_b[:, 1], color='blue', label='Kelas B')
plt.scatter(new_point[0], new_point[1], color='green', s=100, marker='*', label='Titik Baru')

# Hitung jarak dari titik baru ke semua titik
distances_a = [np.linalg.norm(new_point - point) for point in class_a]
distances_b = [np.linalg.norm(new_point - point) for point in class_b]

# Temukan k (misalnya 3) titik terdekat
k = 3
closest_indices = np.argsort(distances_a + distances_b)[:k]
closest_classes = ['A' if i < len(distances_a) else 'B' for i in closest_indices]

# Gambar garis ke k titik terdekat
all_points = np.vstack((class_a, class_b))
for i in closest_indices:
    plt.plot([new_point[0], all_points[i, 0]], [new_point[1], all_points[i, 1]], 'k--')

plt.title(f'K-Nearest Neighbors (k={k}): Kelas yang diprediksi = {max(set(closest_classes), key=closest_classes.count)}')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

## Latihan

1. Buat dua vektor 3D dan hitung:
   - Dot product di antara keduanya
   - Cross product di antara keduanya
   - Sudut di antara kedua vektor (dalam derajat)

2. Implementasikan perhitungan jarak cosinus antara dua vektor, yang didefinisikan sebagai:
   $$d_{cos}(\vec{v}, \vec{w}) = 1 - \frac{\vec{v} \cdot \vec{w}}{||\vec{v}|| \cdot ||\vec{w}||}$$

3. Buat beberapa titik data 2D acak dan implementasikan algoritma k-nearest neighbors sederhana untuk mengklasifikasikan titik baru.

```python
# Latihan 1: Operasi pada vektor 3D
v = np.array([2, 3, 4])
w = np.array([5, 1, 2])

# Dot product
dot_product = np.dot(v, w)
print("Dot product v · w =", dot_product)

# Cross product
cross_product = np.cross(v, w)
print("Cross product v × w =", cross_product)

# Norma vektor
norm_v = np.linalg.norm(v)
norm_w = np.linalg.norm(w)
print("Norma v =", norm_v)
print("Norma w =", norm_w)

# Sudut antara vektor (dalam radian kemudian dikonversi ke derajat)
cos_theta = dot_product / (norm_v * norm_w)
theta_rad = np.arccos(cos_theta)
theta_deg = np.degrees(theta_rad)
print("Cosinus sudut =", cos_theta)
print("Sudut antara v dan w =", theta_deg, "derajat")

# Latihan 2: Jarak cosinus
def cosine_distance(v, w):
    """Menghitung jarak cosinus antara dua vektor."""
    dot_product = np.dot(v, w)
    norm_v = np.linalg.norm(v)
    norm_w = np.linalg.norm(w)
    
    # Menghindari division by zero
    if norm_v == 0 or norm_w == 0:
        return 1.0  # Maximum distance
        
    cosine_similarity = dot_product / (norm_v * norm_w)
    return 1 - cosine_similarity

# Coba dengan beberapa vektor
v1 = np.array([1, 0, 0])  # Vektor di sepanjang sumbu x
v2 = np.array([0, 1, 0])  # Vektor di sepanjang sumbu y
v3 = np.array([1, 1, 0])  # Vektor di antara sumbu x dan y
v4 = np.array([-1, 0, 0])  # Vektor berlawanan dengan v1

print("\nJarak cosinus antara v1 dan v2:", cosine_distance(v1, v2))  # Seharusnya 1 (tegak lurus)
print("Jarak cosinus antara v1 dan v3:", cosine_distance(v1, v3))  # Seharusnya 0.2929 (sudut 45°)
print("Jarak cosinus antara v1 dan v4:", cosine_distance(v1, v4))  # Seharusnya 2 (berlawanan)
print("Jarak cosinus antara v1 dan v1:", cosine_distance(v1, v1))  # Seharusnya 0 (sama)

# Latihan 3: Simple k-NN dengan vektor 2D
# Generate data acak untuk dua kelas
np.random.seed(123)
class1 = np.random.randn(10, 2) + np.array([-3, -3])
class2 = np.random.randn(10, 2) + np.array([3, 3])

# Gabungkan data dan label
X = np.vstack((class1, class2))
y = np.array(['Kelas 1'] * len(class1) + ['Kelas 2'] * len(class2))

def simple_knn(X, y, x_new, k=3):
    """
    Prediksi kelas dari x_new menggunakan k-Nearest Neighbors
    
    Parameter:
    X: array, data training
    y: array, label kelas training
    x_new: array, data yang akan diklasifikasikan
    k: integer, jumlah tetangga terdekat
    
    Return:
    predicted_class: label kelas yang diprediksi
    """
    # Hitung jarak Euclidean
    distances = [np.linalg.norm(x - x_new) for x in X]
    
    # Temukan indeks dari k tetangga terdekat
    nearest_indices = np.argsort(distances)[:k]
    
    # Ambil kelas dari tetangga terdekat
    nearest_classes = y[nearest_indices]
    
    # Voting
    unique_classes, counts = np.unique(nearest_classes, return_counts=True)
    return unique_classes[np.argmax(counts)]

# Titik baru yang akan diklasifikasikan
new_point = np.array([0, 0])

# Prediksi kelas
predicted_class = simple_knn(X, y, new_point, k=5)
print(f"\nPrediksi kelas untuk titik {new_point}: {predicted_class}")

# Visualisasi
plt.figure(figsize=(10, 8))
plt.scatter(class1[:, 0], class1[:, 1], color='red', label='Kelas 1')
plt.scatter(class2[:, 0], class2[:, 1], color='blue', label='Kelas 2')
plt.scatter(new_point[0], new_point[1], color='green', s=100, marker='*', label='Titik Baru')

# Gambar garis ke k titik terdekat
distances = [np.linalg.norm(x - new_point) for x in X]
nearest_indices = np.argsort(distances)[:5]
for idx in nearest_indices:
    plt.plot([new_point[0], X[idx, 0]], [new_point[1], X[idx, 1]], 'k--', alpha=0.5)
    plt.text(X[idx, 0], X[idx, 1], f'#{idx+1}', fontsize=10)

plt.title(f'k-NN (k=5): Prediksi = {predicted_class}')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```
