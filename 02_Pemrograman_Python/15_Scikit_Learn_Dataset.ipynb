Scikit-learn menyediakan beberapa dataset bawaan yang dapat digunakan untuk belajar machine learning. Dataset ini sangat berguna untuk:

Mari kita mulai dengan mengimpor library yang diperlukan:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets

# Untuk tampilan yang lebih baik
# plt.style.use('seaborn')
%matplotlib inline
```

## 2. Jenis-jenis Dataset di Scikit-learn

Scikit-learn menyediakan beberapa dataset populer yang sering digunakan dalam pembelajaran machine learning:

1. **Datasets untuk Klasifikasi**:
   - Iris Dataset (klasifikasi bunga)
   - Digits Dataset (pengenalan angka tulisan tangan)
   - Breast Cancer Dataset (diagnosis kanker payudara)

2. **Datasets untuk Regresi**:
   - Boston House Prices (prediksi harga rumah)
   - Diabetes Dataset (prediksi perkembangan diabetes)

3. **Datasets untuk Clustering**:
   - Make Blobs (dataset sintetis untuk clustering)
   - Make Moons (dataset berbentuk bulan sabit)

Mari kita mulai dengan dataset yang paling populer: Iris Dataset

```python
# Memuat dataset Iris
iris = datasets.load_iris()

# Mengubah data menjadi DataFrame
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

# Menampilkan 5 baris pertama
print("Tampilan dataset Iris:")
print(iris_df.head())

print("\nInformasi dataset:")
print(f"Jumlah sampel: {iris.data.shape[0]}")
print(f"Jumlah fitur: {iris.data.shape[1]}")
print(f"Nama-nama fitur: {', '.join(iris.feature_names)}")
print(f"Nama-nama target: {', '.join(iris.target_names)}")
```

### Visualisasi Dataset Iris

Mari kita buat visualisasi sederhana untuk memahami hubungan antar fitur dalam dataset Iris:

```python
# Membuat scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(iris.data[:, 0], iris.data[:, 1], 
                     c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title('Iris Dataset: Sepal Length vs Sepal Width')
plt.colorbar(scatter)
plt.show()
```

## 3. Dataset untuk Regresi: Boston House Prices

Sekarang mari kita lihat contoh dataset untuk masalah regresi. Dataset Boston House Prices berisi informasi tentang harga rumah di Boston beserta berbagai fitur yang mempengaruhinya.

```python
# Import required libraries
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np

# Load California Housing dataset
california = fetch_california_housing()

# Convert to DataFrame
california_df = pd.DataFrame(california.data, columns=california.feature_names)
california_df['PRICE'] = california.target

print("Tampilan dataset California House Prices:")
print(california_df.head())

print("\nInformasi dataset:")
print(f"Jumlah sampel: {california.data.shape[0]}")
print(f"Jumlah fitur: {california.data.shape[1]}")
print(f"Nama-nama fitur: {', '.join(california.feature_names)}")
```

## 4. Dataset untuk Klasifikasi Digit: Handwritten Digits

Dataset ini berisi gambar angka tulisan tangan dari 0-9. Ini adalah dataset yang sangat populer untuk belajar klasifikasi gambar.

```python
# Memuat dataset digits
digits = datasets.load_digits()

# Menampilkan beberapa contoh gambar
fig, axes = plt.subplots(2, 5, figsize=(10, 4))
for ax, image, label in zip(axes.flat, digits.images[:10], digits.target[:10]):
    ax.imshow(image, cmap='binary')
    ax.set_title(f'Digit: {label}')
    ax.axis('off')
plt.tight_layout()
plt.show()

print("\nInformasi dataset Digits:")
print(f"Jumlah sampel: {digits.data.shape[0]}")
print(f"Dimensi gambar: {digits.images[0].shape[0]}x{digits.images[0].shape[1]}")
print(f"Jumlah kelas: {len(np.unique(digits.target))}")
```

### Solusi Latihan 1: Eksplorasi Dataset Breast Cancer

Berikut adalah solusi untuk latihan pertama:

```python
# Solusi Latihan 1
from sklearn.datasets import load_breast_cancer

# 1. Muat dataset
cancer = load_breast_cancer()

# 2. Ubah ke DataFrame
cancer_df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
cancer_df['target'] = cancer.target

# 3. Tampilkan informasi dasar
print("Informasi Dataset Breast Cancer:")
print(f"Jumlah sampel: {cancer.data.shape[0]}")
print(f"Jumlah fitur: {cancer.data.shape[1]}")
print(f"Nama kelas: {cancer.target_names}")
print("\nDistribusi kelas:")
print(pd.Series(cancer.target).value_counts())

# 4. Visualisasi
plt.figure(figsize=(10, 6))
plt.scatter(cancer.data[:, 0], cancer.data[:, 1], 
           c=cancer.target, cmap='coolwarm')
plt.xlabel(cancer.feature_names[0])
plt.ylabel(cancer.feature_names[1])
plt.title('Breast Cancer Dataset: Feature 0 vs Feature 1')
plt.colorbar(label='Target')
plt.show()
```

### Solusi Latihan 2: Membuat Dataset Sintetis

Berikut adalah solusi untuk latihan kedua:

```python
# Solusi Latihan 2
from sklearn.datasets import make_blobs

# 1. Buat dataset
X, y = make_blobs(n_samples=300, centers=3, cluster_std=1.0,
                 random_state=42)

# 2. Visualisasikan
plt.figure(figsize=(10, 6))
scatter = plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title('Dataset Sintetis dengan 3 Cluster')
plt.colorbar(scatter)
plt.show()

# 3. Eksperimen dengan parameter berbeda
# Membuat dataset dengan cluster yang lebih terpisah
X2, y2 = make_blobs(n_samples=300, centers=3, cluster_std=0.5,
                    random_state=42)

plt.figure(figsize=(10, 6))
scatter = plt.scatter(X2[:, 0], X2[:, 1], c=y2, cmap='viridis')
plt.title('Dataset Sintetis dengan Cluster Lebih Terpisah')
plt.colorbar(scatter)
plt.show()
```
