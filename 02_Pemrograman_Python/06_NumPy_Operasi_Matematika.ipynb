# NumPy Operasi Matematika

NumPy menyediakan berbagai operasi matematika yang sangat penting dalam machine learning. Operasi-operasi ini dioptimasi untuk performa tinggi dan mudah digunakan.

## Jenis-jenis Operasi Matematika

1. **Operasi Aritmatika Dasar**:
   - Penjumlahan (`+`, `add`)
   - Pengurangan (`-`, `subtract`)
   - Perkalian (`*`, `multiply`)
   - Pembagian (`/`, `divide`)
   - Perpangkatan (`**`, `power`)

2. **Operasi Matriks**:
   - Dot product (`dot`)
   - Matrix multiplication (`@`, `matmul`)
   - Transpose (`T`)
   - Inverse (`inv`)
   - Determinant (`det`)

3. **Operasi Statistik**:
   - Mean, Median, Mode
   - Standard deviation, Variance
   - Min, Max
   - Percentiles
   - Correlation

4. **Universal Functions (ufuncs)**:
   - Trigonometric functions (sin, cos, tan)
   - Exponential and logarithmic functions
   - Rounding functions
   - Logical operations

```python
# Import library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Operasi Aritmatika Dasar
print("1. Operasi Aritmatika Dasar:")

# Buat dua array
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])

print("Array a:", a)
print("Array b:", b)
print("\nOperasi dasar:")
print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a ** 2 =", a ** 2)

# Operasi dengan broadcasting
scalar = 2
print("\nOperasi dengan scalar:")
print("a + scalar =", a + scalar)
print("a * scalar =", a * scalar)

# Visualisasi operasi aritmatika
plt.figure(figsize=(15, 5))

# Original arrays
plt.subplot(131)
plt.bar(['a1', 'a2', 'a3', 'a4'], a, alpha=0.5, label='Array a')
plt.bar(['a1', 'a2', 'a3', 'a4'], b, alpha=0.5, label='Array b')
plt.title('Original Arrays')
plt.legend()

# Addition
plt.subplot(132)
plt.bar(['sum1', 'sum2', 'sum3', 'sum4'], a + b)
plt.title('Addition (a + b)')

# Multiplication
plt.subplot(133)
plt.bar(['prod1', 'prod2', 'prod3', 'prod4'], a * b)
plt.title('Multiplication (a * b)')

plt.tight_layout()
plt.show()

# Contoh dalam machine learning: Feature scaling
# Generate sample data
np.random.seed(42)
features = np.random.randn(100, 3)  # 100 samples, 3 features

# Min-Max scaling
min_vals = features.min(axis=0)
max_vals = features.max(axis=0)
features_scaled = (features - min_vals) / (max_vals - min_vals)

print("\nFeature Scaling:")
print("Original range:", [(features[:, i].min(), features[:, i].max()) for i in range(3)])
print("Scaled range:", [(features_scaled[:, i].min(), features_scaled[:, i].max()) for i in range(3)])

# Visualisasi feature scaling
plt.figure(figsize=(12, 4))

plt.subplot(121)
plt.title("Original Features")
sns.boxplot(data=features)
plt.ylabel("Value")

plt.subplot(122)
plt.title("Scaled Features")
sns.boxplot(data=features_scaled)
plt.ylabel("Value")

plt.tight_layout()
plt.show()
```

# 2. Operasi Matriks

NumPy menyediakan berbagai operasi matriks yang sangat penting dalam machine learning dan komputasi numerik. Beberapa operasi matriks yang sering digunakan:

## 2.1 Dot Product dan Matrix Multiplication
- `np.dot(a, b)` atau `a @ b`: Perkalian matriks
- `np.matmul(a, b)`: Perkalian matriks (alternatif)
- `np.inner(a, b)`: Inner product
- `np.outer(a, b)`: Outer product

## 2.2 Operasi Linear Algebra
- `np.linalg.inv(a)`: Invers matriks
- `np.linalg.det(a)`: Determinan matriks
- `np.linalg.eig(a)`: Eigenvalues dan eigenvectors
- `np.linalg.solve(a, b)`: Menyelesaikan sistem persamaan linear

## 2.3 Aplikasi dalam Machine Learning
- **Linear Regression**: Menggunakan dot product untuk menghitung prediksi
- **Neural Networks**: Menggunakan matrix multiplication untuk forward propagation
- **PCA**: Menggunakan eigenvalues dan eigenvectors untuk dimensionality reduction
- **Sistem Rekomendasi**: Menggunakan dot product untuk similarity measures

Mari kita lihat contoh penggunaan operasi matriks dalam konteks machine learning.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create example matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
v = np.array([1, 2])

print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nVector v:")
print(v)

# Matrix multiplication
print("\n1. Matrix Multiplication (A @ B):")
print(np.dot(A, B))  # or A @ B

# Matrix-vector multiplication
print("\n2. Matrix-Vector Multiplication (A @ v):")
print(np.dot(A, v))  # or A @ v

# Inner product of vectors
print("\n3. Inner Product of v with itself:")
print(np.inner(v, v))

# Inverse of matrix A
print("\n4. Inverse of Matrix A:")
print(np.linalg.inv(A))

# Verify inverse multiplication
print("\n5. Verify A @ A^(-1)  I:")
print(np.dot(A, np.linalg.inv(A)))

# Determinant
print("\n6. Determinant of A:")
print(np.linalg.det(A))

# Eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)
print("\n7. Eigenvalues of A:")
print(eigenvalues)
print("\n8. Eigenvectors of A:")
print(eigenvectors)
```

### Understanding the Results

Let's break down what we just computed:

1. **Matrix Multiplication**: We multiplied two 2x2 matrices A and B. This operation is fundamental in neural networks where we multiply weight matrices with input matrices.

2. **Matrix-Vector Multiplication**: We multiplied matrix A with vector v. This is common in transforming data points or computing neural network layers.

3. **Inner Product**: We computed the inner product of vector v with itself, which gives us the squared magnitude of the vector.

4. **Matrix Inverse**: We found A. Notice how when we multiply A with its inverse, we get (approximately) the identity matrix.

5. **Determinant**: The determinant (-2) tells us about the scaling factor of the linear transformation represented by matrix A.

6. **Eigenvalues and Eigenvectors**: These are crucial in many ML applications like PCA and spectral clustering.

Now, let's look at a practical machine learning example: implementing a simple linear regression using these matrix operations.

```python
# Generate some sample data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)  # 100 samples, 1 feature
y = 4 + 3 * X + np.random.randn(100, 1)  # True relationship: y = 4 + 3x + noise

# Add a column of ones to X for the bias term
X_b = np.c_[np.ones((100, 1)), X]

# Compute the parameters using the normal equation:  = (X^T X)^(-1) X^T y
theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

print("Estimated parameters (bias, slope):", theta.flatten())

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.5, label='Data points')
plt.plot(X, X_b.dot(theta), color='red', linewidth=2, label='Linear regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression using Matrix Operations')
plt.legend()
plt.grid(True)
plt.show()
```

### Linear Regression Example Using Matrix Operations

The code above demonstrates a practical application of matrix operations in machine learning through linear regression. Let's break down what's happening:

1. **Data Generation**:
   - We created 100 samples with one feature (X)
   - The true relationship is y = 4 + 3x + random noise
   - This simulates a real-world scenario where data has some linear relationship with noise

2. **Matrix Operations Used**:
   - We added a column of ones to X (X_b) for the bias term using `np.c_[]`
   - The normal equation  = (X^T X)^(-1) X^T y involves several matrix operations:
     - Matrix transpose (X^T)
     - Matrix multiplication (X^T X)
     - Matrix inverse ((X^T X)^(-1))
     - Matrix-vector multiplication with y

3. **Results**:
   - The estimated parameters are:
     - Bias (intercept)  4.22 (true value: 4)
     - Slope  2.77 (true value: 3)
   - The plot shows how well our linear regression line (in red) fits the scattered data points

This example shows how matrix operations are fundamental to implementing machine learning algorithms efficiently. The normal equation gives us a direct solution to find the optimal parameters for linear regression without needing iterative optimization methods.

```python
# Make predictions on new data points
X_new = np.array([[0], [2], [4]])  # Three new data points
X_new_b = np.c_[np.ones((3, 1)), X_new]  # Add bias term
y_pred = X_new_b.dot(theta)

print("Predictions for X = [0, 2, 4]:")
for x, pred in zip(X_new, y_pred):
    print(f"X = {x[0]:.1f}  y = {pred:.2f}")

# Visualize the predictions
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', alpha=0.5, label='Training data')
plt.plot(X, X_b.dot(theta), color='red', label='Regression line')
plt.scatter(X_new, y_pred, color='green', s=100, label='Predictions')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression: Training Data and Predictions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Generate sample data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Add bias term to X
X_b = np.c_[np.ones((100, 1)), X]

# Calculate parameters using normal equation
theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# Make predictions on new data points
X_new = np.array([[0], [2], [4]])  # Three new data points
X_new_b = np.c_[np.ones((3, 1)), X_new]  # Add bias term
y_pred = X_new_b.dot(theta)

print("Predictions for X = [0, 2, 4]:")
for i in range(len(X_new)):
    print(f"X = {X_new[i][0]:.1f}  y = {y_pred[i][0]:.2f}")

# Visualize the predictions
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', alpha=0.5, label='Training data')
plt.plot(X, X_b.dot(theta), color='red', label='Regression line')
plt.scatter(X_new, y_pred, color='green', s=100, label='Predictions')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression: Training Data and Predictions')
plt.legend()
plt.grid(True)
plt.show()
```

# Model Evaluation

Now that we have our trained model and have made some predictions, let's evaluate its performance using common metrics:

1. Mean Squared Error (MSE)
2. Root Mean Squared Error (RMSE)
3. R-squared (R) score

These metrics will help us understand how well our model fits the data.

```python
# Calculate predictions for all training data
y_pred_train = X_b.dot(theta)

# Calculate MSE
mse = np.mean((y - y_pred_train) ** 2)

# Calculate RMSE
rmse = np.sqrt(mse)

# Calculate R-squared
y_mean = np.mean(y)
total_sum_squares = np.sum((y - y_mean) ** 2)
residual_sum_squares = np.sum((y - y_pred_train) ** 2)
r2 = 1 - (residual_sum_squares / total_sum_squares)

print("Model Performance Metrics:")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared (R): {r2:.4f}")

# Visualize residuals
plt.figure(figsize=(10, 6))
plt.scatter(y_pred_train, y - y_pred_train, c='blue', alpha=0.5)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.axhline(y=0, color='r', linestyle='--')
plt.grid(True)
plt.show()
```

```python
# Reshape predictions to match y shape
y_pred_train = y_pred_train.reshape(-1, 1)

# Calculate residuals
residuals = y - y_pred_train

# Visualize residuals
plt.figure(figsize=(10, 6))
plt.scatter(y_pred_train.flatten(), residuals.flatten(), c='blue', alpha=0.5)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.axhline(y=0, color='r', linestyle='--')
plt.grid(True)
plt.show()

print("\nModel Performance Analysis:")
print("-" * 50)
print(f"Mean Squared Error (MSE): {mse:.4f}")
print("- This represents the average squared difference between predictions and actual values")
print(f"\nRoot Mean Squared Error (RMSE): {rmse:.4f}")
print("- This is the square root of MSE, giving us an error measure in the same units as our target variable")
print(f"\nR-squared (R): {r2:.4f}")
print("- This indicates the proportion of variance in the dependent variable explained by the model")
print("- A negative R suggests our model performs worse than a horizontal line")
```

## Next Steps for Improvement

2. Consider adding polynomial features

```python
# Add polynomial features
X_poly = np.c_[X, X**2]
X_poly_b = np.c_[np.ones((100, 1)), X_poly]  # Add bias term

# Calculate parameters using normal equation for polynomial features
theta_poly = np.linalg.inv(X_poly_b.T.dot(X_poly_b)).dot(X_poly_b.T).dot(y)

# Make predictions with polynomial features
y_pred_poly = X_poly_b.dot(theta_poly)

# Calculate metrics for polynomial model
mse_poly = np.mean((y - y_pred_poly) ** 2)
rmse_poly = np.sqrt(mse_poly)
r2_poly = 1 - (np.sum((y - y_pred_poly) ** 2) / np.sum((y - np.mean(y)) ** 2))

print("Polynomial Model Performance:")
print("-" * 50)
print(f"Mean Squared Error (MSE): {mse_poly:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse_poly:.4f}")
print(f"R-squared (R): {r2_poly:.4f}")

# Visualize the polynomial fit
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', alpha=0.5, label='Training data')

# Sort X for smooth curve plotting
X_sorted = np.sort(X, axis=0)
X_poly_sorted = np.c_[X_sorted, X_sorted**2]
X_poly_b_sorted = np.c_[np.ones((100, 1)), X_poly_sorted]
y_poly_sorted = X_poly_b_sorted.dot(theta_poly)

plt.plot(X_sorted, y_poly_sorted, color='red', label='Polynomial regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression: Training Data and Model Fit')
plt.legend()
plt.grid(True)
plt.show()
```
