# Struktur Dasar Kondisi di Python

Pada notebook ini, kita akan membahas struktur kondisional dalam Python yang sering digunakan dalam machine learning.

## Apa itu Struktur Kondisional?

Struktur kondisional adalah pernyataan yang memungkinkan program untuk membuat keputusan berdasarkan kondisi tertentu. Python menyediakan beberapa jenis struktur kondisional:
1. `if` statement
2. `if-else` statement
3. `if-elif-else` statement
4. Operator ternary

## Pentingnya Kondisional dalam Machine Learning

Dalam machine learning, struktur kondisional digunakan untuk:
- Pemilihan model berdasarkan karakteristik data
- Penanganan kasus khusus dalam data
- Early stopping dalam training
- Pemilihan hyperparameter
- Validasi input dan output
- Penanganan error dan exception

````python
# Import library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt

# Contoh sederhana if statement
nilai = 75
if nilai >= 70:
    print("Lulus!")

# Contoh if-else
accuracy = 0.85
if accuracy > 0.8:
    print("Model memiliki performa baik")
else:
    print("Model perlu perbaikan")

# Contoh if-elif-else
loss = 0.3
if loss < 0.2:
    print("Model sangat baik")
elif loss < 0.4:
    print("Model cukup baik")
else:
    print("Model perlu perbaikan")
````

## Operator Perbandingan

Python menyediakan berbagai operator perbandingan yang dapat digunakan dalam kondisi:

- `==` : sama dengan
- `!=` : tidak sama dengan
- `<`  : kurang dari
- `>`  : lebih dari
- `<=` : kurang dari atau sama dengan
- `>=` : lebih dari atau sama dengan
- `is` : identitas objek sama
- `is not` : identitas objek berbeda

Mari kita lihat contoh penggunaan operator perbandingan dalam konteks machine learning:

````python
# Contoh penggunaan operator perbandingan dalam ML

# Simulasi evaluasi model
accuracy = 0.92
precision = 0.85
recall = 0.88
threshold = 0.9

print("Evaluasi Metrik Model:")

# Menggunakan operator perbandingan
if accuracy >= threshold:
    print("✓ Accuracy mencapai threshold")
else:
    print("✗ Accuracy di bawah threshold")

if precision >= threshold:
    print("✓ Precision mencapai threshold")
else:
    print("✗ Precision di bawah threshold")

if recall >= threshold:
    print("✓ Recall mencapai threshold")
else:
    print("✗ Recall di bawah threshold")

# Menggunakan multiple conditions
if accuracy >= threshold and precision >= threshold and recall >= threshold:
    print("\nModel siap untuk production!")
else:
    print("\nModel masih perlu perbaikan.")

# Contoh penggunaan is dan is not
import numpy as np

# Membuat array numpy
array1 = np.array([1, 2, 3])
array2 = np.array([1, 2, 3])

print("\nPerbandingan array:")
print("array1 == array2:", array1 == array2)  # Membandingkan nilai
print("array1 is array2:", array1 is array2)  # Membandingkan identitas objek
````

## Operator Logika

Python menyediakan tiga operator logika dasar:
- `and` : kedua kondisi harus benar
- `or`  : salah satu kondisi harus benar
- `not` : membalikkan nilai boolean

Operator logika sangat berguna dalam machine learning untuk membuat kondisi kompleks. Berikut beberapa contoh penggunaannya:

````python
# Contoh penggunaan operator logika dalam ML

# Simulasi parameter model
learning_rate = 0.01
batch_size = 32
epochs = 100
early_stopping_patience = 5

# Validasi parameter model
is_valid_lr = 0.0001 <= learning_rate <= 0.1
is_valid_batch = batch_size in [16, 32, 64, 128]
is_valid_epochs = epochs > 0 and epochs <= 1000

print("Validasi Parameter Model:")
print(f"Learning rate valid: {is_valid_lr}")
print(f"Batch size valid: {is_valid_batch}")
print(f"Epochs valid: {is_valid_epochs}")

# Menggunakan operator and
if is_valid_lr and is_valid_batch and is_valid_epochs:
    print("\nSemua parameter valid, memulai training...")
else:
    print("\nAda parameter yang tidak valid!")

# Contoh penggunaan or untuk pengecekan kondisi error
loss_increasing = True
val_loss_increasing = True
patience_exceeded = False

if (loss_increasing and val_loss_increasing) or patience_exceeded:
    print("\nMenghentikan training karena:")
    if loss_increasing and val_loss_increasing:
        print("- Loss dan validation loss meningkat")
    if patience_exceeded:
        print("- Patience exceeded")

# Contoh penggunaan not
is_converged = False
if not is_converged:
    print("\nModel belum konvergen, melanjutkan training...")
````

## Operator Ternary

Python menyediakan cara singkat untuk menulis kondisi if-else dalam satu baris, yang disebut operator ternary. Formatnya adalah:

```python
value_if_true if condition else value_if_false
```

Operator ternary sangat berguna untuk membuat kode lebih ringkas dan mudah dibaca. Berikut beberapa contoh penggunaannya dalam konteks machine learning:

````python
# Contoh penggunaan operator ternary

# Klasifikasi sederhana
def classify_sample(value, threshold=0.5):
    return "Positif" if value >= threshold else "Negatif"

# Contoh penggunaan
samples = [0.2, 0.6, 0.45, 0.8, 0.3]
predictions = [classify_sample(x) for x in samples]

print("Klasifikasi sampel:")
for sample, pred in zip(samples, predictions):
    print(f"Nilai: {sample:.2f} -> Kelas: {pred}")

# Penggunaan ternary untuk pemilihan model
accuracy = 0.85
model_status = "production ready" if accuracy > 0.8 else "needs improvement"
print(f"\nStatus model: {model_status}")

# Penggunaan ternary untuk pemilihan hyperparameter
batch_size = 32
learning_rate = 0.01 if batch_size >= 32 else 0.001
print(f"Learning rate yang dipilih: {learning_rate}")

# Penggunaan ternary dengan multiple conditions
val_loss = 0.3
epochs = 50
training_status = ("early stopping" if val_loss < 0.2 else 
                  "continue" if epochs < 100 else 
                  "finished")
print(f"Status training: {training_status}")
````

## Nested Conditions

Kondisi bersarang (nested conditions) terjadi ketika kita menempatkan satu atau lebih kondisi di dalam kondisi lain. Meskipun berguna, penggunaan nested conditions yang terlalu dalam dapat membuat kode sulit dibaca.

Berikut contoh penggunaan nested conditions dalam konteks machine learning:

````python
# Contoh nested conditions dalam ML

def evaluate_model_performance(accuracy, loss, val_accuracy, val_loss):
    if accuracy > 0.8:
        if val_accuracy > 0.8:
            if loss < 0.2:
                if val_loss < 0.2:
                    return "Model Excellent"
                else:
                    return "Model Good, but validation loss high"
            else:
                return "Model Good, but training loss high"
        else:
            return "Model might be overfitting"
    else:
        return "Model needs improvement"

# Test beberapa skenario
scenarios = [
    (0.90, 0.15, 0.85, 0.18),  # Skenario 1
    (0.85, 0.18, 0.75, 0.25),  # Skenario 2
    (0.75, 0.30, 0.70, 0.35)   # Skenario 3
]

print("Evaluasi Model dalam Berbagai Skenario:")
for i, (acc, loss, val_acc, val_loss) in enumerate(scenarios, 1):
    result = evaluate_model_performance(acc, loss, val_acc, val_loss)
    print(f"\nSkenario {i}:")
    print(f"Accuracy: {acc:.2f}, Loss: {loss:.2f}")
    print(f"Val Accuracy: {val_acc:.2f}, Val Loss: {val_loss:.2f}")
    print(f"Hasil Evaluasi: {result}")

# Alternatif yang lebih baik menggunakan early returns
def evaluate_model_performance_better(accuracy, loss, val_accuracy, val_loss):
    if accuracy <= 0.8:
        return "Model needs improvement"
    
    if val_accuracy <= 0.8:
        return "Model might be overfitting"
        
    if loss >= 0.2:
        return "Model Good, but training loss high"
        
    if val_loss >= 0.2:
        return "Model Good, but validation loss high"
        
    return "Model Excellent"

print("\nEvaluasi dengan kode yang lebih bersih:")
for i, scenario in enumerate(scenarios, 1):
    result = evaluate_model_performance_better(*scenario)
    print(f"Skenario {i}: {result}")
````

## Latihan

1. Buatlah fungsi untuk memilih algoritma machine learning berdasarkan karakteristik dataset (jumlah sampel, jumlah fitur, tipe masalah).
2. Implementasikan sistem evaluasi model yang mempertimbangkan multiple metrics (accuracy, precision, recall, F1-score).
3. Buat fungsi untuk memilih hyperparameter optimal berdasarkan beberapa kondisi.

Mari kita selesaikan latihan pertama sebagai contoh:

````python
# Latihan 1: Pemilihan Algoritma ML

def choose_algorithm(n_samples, n_features, problem_type, is_linear=None):
    """
    Memilih algoritma ML berdasarkan karakteristik dataset
    
    Parameters:
    -----------
    n_samples : int
        Jumlah sampel dalam dataset
    n_features : int
        Jumlah fitur dalam dataset
    problem_type : str
        Tipe masalah ('classification' atau 'regression')
    is_linear : bool, optional
        Apakah data bersifat linear
    
    Returns:
    --------
    str
        Nama algoritma yang direkomendasikan
    """
    
    if problem_type == 'classification':
        if n_samples < 1000:
            if n_features < 50:
                return "K-Nearest Neighbors"
            else:
                return "Random Forest"
        else:
            if is_linear:
                return "Logistic Regression"
            else:
                return "Gradient Boosting"
    
    elif problem_type == 'regression':
        if n_samples < 1000:
            if is_linear:
                return "Linear Regression"
            else:
                return "Decision Tree"
        else:
            if is_linear:
                return "Ridge Regression"
            else:
                return "XGBoost"
    
    else:
        return "Unsupported problem type"

# Test fungsi dengan berbagai skenario
test_cases = [
    (500, 20, 'classification', True),
    (1500, 100, 'classification', False),
    (800, 10, 'regression', True),
    (2000, 50, 'regression', False)
]

print("Rekomendasi Algoritma untuk Berbagai Skenario:")
for i, (samples, features, prob_type, linear) in enumerate(test_cases, 1):
    algo = choose_algorithm(samples, features, prob_type, linear)
    print(f"\nSkenario {i}:")
    print(f"Jumlah sampel: {samples}")
    print(f"Jumlah fitur: {features}")
    print(f"Tipe masalah: {prob_type}")
    print(f"Linear: {linear}")
    print(f"Algoritma yang direkomendasikan: {algo}")

# Visualisasi decision boundary
import numpy as np
import matplotlib.pyplot as plt

# Membuat plot untuk visualisasi pemilihan algoritma berdasarkan jumlah sampel dan fitur
n_samples = np.linspace(100, 2000, 100)
n_features = np.linspace(1, 100, 100)
X, Y = np.meshgrid(n_samples, n_features)

plt.figure(figsize=(12, 6))

# Plot untuk klasifikasi
plt.subplot(1, 2, 1)
plt.title('Pemilihan Algoritma Klasifikasi')
plt.axvline(x=1000, color='r', linestyle='--', label='Sample threshold')
plt.axhline(y=50, color='g', linestyle='--', label='Feature threshold')
plt.xlabel('Jumlah Sampel')
plt.ylabel('Jumlah Fitur')
plt.legend()
plt.grid(True)

# Plot untuk regresi
plt.subplot(1, 2, 2)
plt.title('Pemilihan Algoritma Regresi')
plt.axvline(x=1000, color='r', linestyle='--', label='Sample threshold')
plt.xlabel('Jumlah Sampel')
plt.ylabel('Jumlah Fitur')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
````
