```python
# Import library yang diperlukan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Membuat data contoh
np.random.seed(42)
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x/5)

# Data untuk scatter plot
n_points = 50
x_scatter = np.random.normal(5, 2, n_points)
y_scatter = np.random.normal(5, 2, n_points)

# Data untuk bar plot
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 50, len(categories))

print("Data telah disiapkan!")
```

## 1. Line Plot

Line plot adalah jenis visualisasi yang paling dasar dan sering digunakan untuk menampilkan tren data dalam bentuk garis. Mari kita lihat beberapa contoh:

```python
# Membuat line plot sederhana
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-')
plt.plot(x, y2, label='cos(x)', color='red', linestyle='--')
plt.plot(x, y3, label='exp(-x/5)', color='green', linestyle=':')

# Menambahkan judul dan label
plt.title('Contoh Line Plot', fontsize=14)
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)

# Menambahkan grid
plt.grid(True, linestyle='--', alpha=0.7)

# Menambahkan legend
plt.legend(fontsize=10)

# Menampilkan plot
plt.show()
```

## 2. Scatter Plot

Scatter plot sangat berguna untuk menampilkan hubungan antara dua variabel. Mari kita buat beberapa contoh scatter plot:

```python
# Membuat scatter plot dengan ukuran dan warna yang bervariasi
plt.figure(figsize=(10, 6))

# Membuat data untuk warna dan ukuran
colors = np.random.rand(n_points)
sizes = np.random.randint(50, 200, n_points)

# Membuat scatter plot
plt.scatter(x_scatter, y_scatter, c=colors, s=sizes, alpha=0.6, cmap='viridis')

# Menambahkan colorbar
plt.colorbar(label='Random Values')

# Menambahkan judul dan label
plt.title('Scatter Plot dengan Variasi Ukuran dan Warna', fontsize=14)
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)

# Menambahkan grid
plt.grid(True, linestyle='--', alpha=0.3)

plt.show()
```

## 3. Bar Plot

Bar plot sangat efektif untuk membandingkan nilai antar kategori. Mari kita lihat beberapa variasi bar plot:

```python
# Membuat figure dengan dua subplot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# 1. Bar plot vertikal
ax1.bar(categories, values, color='skyblue', edgecolor='black')
ax1.set_title('Bar Plot Vertikal')
ax1.set_xlabel('Kategori')
ax1.set_ylabel('Nilai')
ax1.grid(True, linestyle='--', alpha=0.3)

# 2. Bar plot horizontal
ax2.barh(categories, values, color='lightgreen', edgecolor='black')
ax2.set_title('Bar Plot Horizontal')
ax2.set_xlabel('Nilai')
ax2.set_ylabel('Kategori')
ax2.grid(True, linestyle='--', alpha=0.3)

# Mengatur layout
plt.tight_layout()
plt.show()
```

## 4. Histogram dan Pie Chart

Histogram digunakan untuk menampilkan distribusi data, sedangkan pie chart untuk menampilkan proporsi dari keseluruhan:

```python
# Membuat figure dengan dua subplot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# 1. Histogram
data_hist = np.random.normal(100, 15, 1000)
ax1.hist(data_hist, bins=30, color='purple', alpha=0.7, edgecolor='black')
ax1.set_title('Histogram')
ax1.set_xlabel('Nilai')
ax1.set_ylabel('Frekuensi')
ax1.grid(True, linestyle='--', alpha=0.3)

# 2. Pie Chart
sizes = [30, 20, 25, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']
colors = ['lightblue', 'lightgreen', 'lightcoral', 'wheat', 'pink']
explode = (0.1, 0, 0, 0, 0)  # memisahkan slice pertama

ax2.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
ax2.set_title('Pie Chart')

# Mengatur layout
plt.tight_layout()
plt.show()
```

## 5. Subplot dan Grid Layout

Matplotlib memungkinkan kita untuk membuat multiple plots dalam satu figure. Mari kita lihat beberapa cara untuk mengatur layout:

```python
# Membuat grid layout dengan 2x2 subplot
fig = plt.figure(figsize=(12, 10))

# Subplot 1: Line plot
ax1 = plt.subplot(2, 2, 1)
ax1.plot(x, y1, 'b-', label='sin(x)')
ax1.set_title('Line Plot')
ax1.legend()
ax1.grid(True)

# Subplot 2: Scatter plot
ax2 = plt.subplot(2, 2, 2)
ax2.scatter(x_scatter, y_scatter, c='red', alpha=0.6)
ax2.set_title('Scatter Plot')
ax2.grid(True)

# Subplot 3: Bar plot
ax3 = plt.subplot(2, 2, 3)
ax3.bar(categories, values, color='green')
ax3.set_title('Bar Plot')
ax3.grid(True)

# Subplot 4: Histogram
ax4 = plt.subplot(2, 2, 4)
ax4.hist(data_hist, bins=20, color='purple', alpha=0.7)
ax4.set_title('Histogram')
ax4.grid(True)

# Mengatur layout
plt.tight_layout()
plt.show()
```

## 6. Studi Kasus: Analisis Data Penjualan

Mari kita terapkan berbagai jenis visualisasi dalam konteks analisis data penjualan:

```python
# Membuat data penjualan
np.random.seed(42)
dates = pd.date_range(start='2024-01-01', periods=30, freq='D')
products = ['Produk A', 'Produk B', 'Produk C']
sales_data = {
    'Tanggal': np.repeat(dates, len(products)),
    'Produk': products * len(dates),
    'Penjualan': np.random.randint(50, 200, len(dates) * len(products)),
    'Revenue': np.random.uniform(1000, 5000, len(dates) * len(products))
}

df_sales = pd.DataFrame(sales_data)

# Membuat visualisasi komprehensif
fig = plt.figure(figsize=(15, 10))

# 1. Time series plot penjualan harian
ax1 = plt.subplot(2, 2, 1)
for product in products:
    product_data = df_sales[df_sales['Produk'] == product]
    ax1.plot(product_data['Tanggal'], product_data['Penjualan'], 
             label=product, marker='o', markersize=4)
ax1.set_title('Trend Penjualan Harian')
ax1.set_xlabel('Tanggal')
ax1.set_ylabel('Jumlah Penjualan')
ax1.legend()
ax1.grid(True)
plt.xticks(rotation=45)

# 2. Bar plot total penjualan per produk
ax2 = plt.subplot(2, 2, 2)
total_sales = df_sales.groupby('Produk')['Penjualan'].sum()
ax2.bar(total_sales.index, total_sales.values, color=['skyblue', 'lightgreen', 'lightcoral'])
ax2.set_title('Total Penjualan per Produk')
ax2.set_ylabel('Total Penjualan')
ax2.grid(True)

# 3. Scatter plot hubungan penjualan dan revenue
ax3 = plt.subplot(2, 2, 3)
colors = {'Produk A': 'red', 'Produk B': 'blue', 'Produk C': 'green'}
for product in products:
    product_data = df_sales[df_sales['Produk'] == product]
    ax3.scatter(product_data['Penjualan'], product_data['Revenue'], 
                label=product, color=colors[product], alpha=0.6)
ax3.set_title('Hubungan Penjualan dan Revenue')
ax3.set_xlabel('Jumlah Penjualan')
ax3.set_ylabel('Revenue')
ax3.legend()
ax3.grid(True)

# 4. Pie chart proporsi revenue
ax4 = plt.subplot(2, 2, 4)
revenue_share = df_sales.groupby('Produk')['Revenue'].sum()
ax4.pie(revenue_share, labels=revenue_share.index, autopct='%1.1f%%',
        colors=['lightblue', 'lightgreen', 'lightcoral'])
ax4.set_title('Proporsi Revenue per Produk')

plt.tight_layout()
plt.show()

# Menampilkan ringkasan statistik
print("\nRingkasan Statistik Penjualan:")
print(df_sales.groupby('Produk').agg({
    'Penjualan': ['mean', 'min', 'max', 'sum'],
    'Revenue': ['mean', 'min', 'max', 'sum']
}).round(2))
```

## 7. Latihan

Sekarang saatnya untuk berlatih! Gunakan dataset yang telah kita buat untuk menyelesaikan soal-soal berikut:

1. Buat line plot yang menampilkan moving average (rata-rata bergerak) 7 hari untuk setiap produk
2. Buat stacked bar chart yang menunjukkan kontribusi setiap produk terhadap total penjualan harian
3. Buat histogram yang membandingkan distribusi revenue untuk setiap produk
4. Buat scatter plot matrix yang menunjukkan hubungan antara Penjualan dan Revenue untuk setiap produk
5. Buat visualisasi yang menunjukkan performa penjualan mingguan

Berikut adalah template kode untuk memulai latihan:

```python
# Template untuk latihan

# Soal 1: Moving Average Plot
# Hint: Gunakan pandas rolling() untuk menghitung moving average

# Soal 2: Stacked Bar Chart
# Hint: Gunakan plt.bar() dengan parameter bottom

# Soal 3: Histogram Perbandingan
# Hint: Gunakan plt.hist() dengan parameter alpha untuk transparansi

# Soal 4: Scatter Plot Matrix
# Hint: Buat subplot grid untuk setiap kombinasi variabel

# Soal 5: Visualisasi Mingguan
# Hint: Gunakan resample('W') untuk data mingguan
```

### Solusi Latihan

Berikut adalah solusi untuk latihan di atas. Cobalah untuk mengerjakan sendiri terlebih dahulu sebelum melihat solusi!

```python
# Solusi Soal 1: Moving Average Plot
plt.figure(figsize=(12, 6))

for product in products:
    product_data = df_sales[df_sales['Produk'] == product]
    # Menghitung moving average 7 hari
    ma = product_data.set_index('Tanggal')['Penjualan'].rolling(window=7).mean()
    plt.plot(ma.index, ma.values, label=f'{product} (MA-7)', linewidth=2)
    # Plot data asli dengan transparansi
    plt.plot(product_data['Tanggal'], product_data['Penjualan'], 
             alpha=0.3, linestyle=':', label=f'{product} (Raw)')

plt.title('7-Day Moving Average Penjualan')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Penjualan')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Solusi Soal 2: Stacked Bar Chart
plt.figure(figsize=(12, 6))

# Pivot data untuk stacked bar chart
daily_sales = df_sales.pivot(index='Tanggal', columns='Produk', values='Penjualan')

# Plot stacked bar
bottom = np.zeros(len(daily_sales))
colors = ['skyblue', 'lightgreen', 'lightcoral']

for i, product in enumerate(products):
    plt.bar(daily_sales.index, daily_sales[product], bottom=bottom, 
            label=product, color=colors[i])
    bottom += daily_sales[product]

plt.title('Kontribusi Penjualan Harian per Produk')
plt.xlabel('Tanggal')
plt.ylabel('Total Penjualan')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Solusi Soal 3: Histogram Perbandingan
plt.figure(figsize=(12, 6))

colors = ['skyblue', 'lightgreen', 'lightcoral']
for i, product in enumerate(products):
    product_data = df_sales[df_sales['Produk'] == product]
    plt.hist(product_data['Revenue'], bins=20, alpha=0.5, 
             label=product, color=colors[i])

plt.title('Distribusi Revenue per Produk')
plt.xlabel('Revenue')
plt.ylabel('Frekuensi')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Solusi Soal 4: Scatter Plot Matrix
fig, axes = plt.subplots(len(products), len(products), figsize=(15, 15))
fig.suptitle('Scatter Plot Matrix: Penjualan vs Revenue', y=1.02)

for i, product1 in enumerate(products):
    for j, product2 in enumerate(products):
        data1 = df_sales[df_sales['Produk'] == product1]
        data2 = df_sales[df_sales['Produk'] == product2]
        
        axes[i, j].scatter(data1['Penjualan'], data2['Revenue'], alpha=0.5)
        axes[i, j].set_xlabel(f'{product1} Penjualan')
        axes[i, j].set_ylabel(f'{product2} Revenue')
        axes[i, j].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Solusi Soal 5: Visualisasi Mingguan
# Menghitung data mingguan
weekly_data = df_sales.set_index('Tanggal').groupby(['Produk']).resample('W')['Penjualan'].sum().reset_index()

plt.figure(figsize=(12, 6))

for product in products:
    product_data = weekly_data[weekly_data['Produk'] == product]
    plt.plot(product_data['Tanggal'], product_data['Penjualan'], 
             marker='o', label=product, linewidth=2)

plt.title('Performa Penjualan Mingguan')
plt.xlabel('Minggu')
plt.ylabel('Total Penjualan')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Menampilkan ringkasan statistik mingguan
print("\nRingkasan Statistik Mingguan:")
weekly_stats = weekly_data.groupby('Produk').agg({
    'Penjualan': ['mean', 'min', 'max']
}).round(2)
print(weekly_stats)
```
